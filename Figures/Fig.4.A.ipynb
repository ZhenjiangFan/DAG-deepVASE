```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

validAssociations = pd.read_csv('Number of validated associations between BMI factors and BMI.csv',index_col=0).T;
print(validAssociations.shape);
validAssociations
```

```python
nutrientIntake = validAssociations['nutrient intake']
bacteriaGenera = validAssociations['bacteria genera']
labels = validAssociations.index;

ind = np.arange(len(nutrientIntake))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, nutrientIntake, width,label='nutrient intake')
rects2 = ax.bar(ind + width/2, bacteriaGenera, width, label='bacteria genera')

ax.yaxis.grid(True)
ax.set_ylabel('# validated associations between BMI factors and BMI')
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(labels, rotation=50, fontsize=10)
ax.legend()

# plt.savefig("Number of validated associations between BMI factors and BMI.svg")

plt.show()
```
