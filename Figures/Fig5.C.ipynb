```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


linears = (6, 9, 2)
nonlinears = (2, 34, 13)

ind = np.arange(len(linears))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, linears, width,label='linear',color='blue')
rects2 = ax.bar(ind + width/2, nonlinears, width, label='nonlinear',color='red')

ax.yaxis.grid(True)
ax.set_ylabel('causal relations learned by DAG-deepVASE')
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(('between genes','from a gene to a clinical feature','betweenclinical features')
                   , rotation=50, fontsize=10)
ax.legend()

plt.show()
```
