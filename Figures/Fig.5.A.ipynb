```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

validAssociations = pd.read_csv('Number of validated associations to PAM50 - molecular-clinical.csv',index_col=0).T;
print(validAssociations.shape);
validAssociations
```

```python
moleculars = validAssociations['moleculars']
clinicals = validAssociations['clinicals']
labels = validAssociations.index;

ind = np.arange(len(moleculars))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, moleculars, width,label='molecular')
rects2 = ax.bar(ind + width/2, clinicals, width, label='clinical')

ax.yaxis.grid(True)
ax.set_ylabel('# validated associations to PAM50')
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(labels, rotation=50, fontsize=10)
ax.legend()

# plt.savefig("Number of validated associations to PAM50 - molecular-clinical.svg")

plt.show()
```
