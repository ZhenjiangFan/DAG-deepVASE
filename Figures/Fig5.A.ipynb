```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


moleculars = (3, 5, 9)
clinicals = (3, 0, 0)
overlaps = (0, 1, 5)

ind = np.arange(len(moleculars))
width = 0.3

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width, moleculars, width,label='genes to PAM50 (TP)')
rects2 = ax.bar(ind, clinicals, width, label='PAM50 to genes (FP)')
rects3 = ax.bar(ind + width, overlaps, width, label='b/w PAM50 to clinical features',color='gray')

ax.yaxis.grid(True)
ax.set_ylabel('# validated associations to PAM50', fontsize=15)
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(('causalMGM','direct-DG','DAG-deepVASE')
                   , rotation=50, fontsize=20)
ax.legend()


plt.show()
```
