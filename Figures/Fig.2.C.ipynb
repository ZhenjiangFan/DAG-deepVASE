```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

predictionPowers = pd.read_csv('Average prediction power on true associations (using 50 repetitions of data).csv',index_col=0);
print(predictionPowers.shape);
predictionPowers
```

```python
DAGDeepVASE = predictionPowers['DAG-DeepVASE'].values;
causalMGM = predictionPowers['causalMGM'].values;
DirectDG = predictionPowers['Direct-DG'].values;
indeces = predictionPowers.index;

ind = np.arange(len(DAGDeepVASE));
width = 0.3;

fig, ax = plt.subplots();
rects1 = ax.bar(ind - width, DAGDeepVASE, width-0.05,label='DAG-DeepVASE',color='red');
rects2 = ax.bar(ind, causalMGM, width-0.05, label='causalMGM',color='gray');
rects3 = ax.bar(ind + width, DirectDG, width-0.05, label='linear-DG',color='gold');


ax.set_ylabel('power', fontsize=15);
ax.set_xticks(range(0,len(indeces)));
ax.set_xticklabels(indeces, rotation=50, fontsize=20);
ax.legend();

# plt.savefig("Average prediction power on true associations (using 50 repetitions of data).svg");

plt.show();
```
