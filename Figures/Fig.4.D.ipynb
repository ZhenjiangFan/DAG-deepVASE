```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

causalities = pd.read_csv('Number of true positives and false positives causalities on BMI.csv',index_col=0).T;
print(causalities.shape);
causalities
```

```python
truePositive = causalities['true positive']
falsePositive = causalities['false positive']
labels = causalities.index;

ind = np.arange(len(truePositive))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, truePositive, width,label='true positive',color='darkblue')
rects2 = ax.bar(ind + width/2, falsePositive, width, label='false positive',color='red')

ax.yaxis.grid(True)
ax.set_ylabel('# causalities')
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(labels, rotation=50, fontsize=10)
ax.legend()

# plt.savefig("Number of true positives and false positives causalities on BMI.svg")

plt.show()
```
