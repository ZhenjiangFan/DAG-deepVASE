```python
import pandas as pd
import numpy as np
import os
```

```python
sample_num = 1000;
feature_num = 200;
#Create column names
col_names = [];
for index in range(0,feature_num):
    name = "F"+str(index+1);
    col_names.append(name);
```

```python
#Load the X data and Y data
X_data = pd.read_csv("data/X_n1000_p200_rep20.csv",names=col_names);
Y_data = pd.read_csv("data/y_si_n1000_p200_rep20.csv",names=["Y"]);

dataset = pd.concat([X_data, Y_data], axis=1, join='inner');
# print(dataset.shape);
dataset
```

```python
import random
selected_nums = random.sample(range(0,1000), sample_num);
selected_nums.sort();
# pd.DataFrame({"RNs":selected_nums}).to_csv("randomly_selected_rows_5.csv");
```

```python
#Generate input data for MGM.
# dataset = dataset.iloc[selected_nums];
dataset.to_csv("data/200p_"+str(sample_num)+"samples_XY.txt", index=None, sep="\t");
```

```python
# X_data = X_data.iloc[selected_nums];
X_data.to_csv("data/200p_"+str(sample_num)+"samples_X.csv", index=None, sep=",");
```
